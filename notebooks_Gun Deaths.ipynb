```python
# do calculations

gfi = [x["GFI"] for x in numbers]
rate = [x["RATE"] for x in numbers]

# do a simple linear regression fitting
slope, intercept, r_value, p_value, std_err = stats.linregress(gfi, rate)

# Pearson correlation coefficient
correlation_coefficient = np.corrcoef(gfi, rate)[0][1]
```

```python
# Plot the numbers

fig, ax = plt.subplots()

# Map each onto a scatterplot we'll create with Matplotlib
#ax.scatter(x=gfi, y=rate)
ax.set(title="Gun death rate in 2020 v. Gun friendly Index")
ax.plot(gfi, rate, 'ro')
ax.axline(xy1=(0, intercept), slope=slope, linestyle="--", color="k")

ax.set_ylabel('Annual Gun deaths / 100K')
ax.set_xlabel('Gun friendly index')

for state in numbers:
    ax.annotate(state["STATE"], (state["GFI"], state["RATE"]))

plt.show()

print(f"P-Value is {p_value}: result is statistically significant.")
print(f"Pearson correlation coefficient is {correlation_coefficient}: a positive relationship exists.")
```
